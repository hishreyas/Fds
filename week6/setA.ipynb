```python
#A1
import numpy as np
import matplotlib.pyplot as plt
x = np.random.randn(50)
y = np.random.randn(50)

x[49]=15
x[48]=20
y[49]=10
y[48]=5



plt.boxplot(x, vert=False)
plt.title('Box-plot')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#A2
import numpy as np
import matplotlib.pyplot as plt
x = np.random.randn(50)
y = np.random.randn(50)


plt.title('Scatter chart')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
colors = 50*np.random.randn(50)
plt.scatter(x,y,c=colors,marker ='*')
plt.show()



plt.plot(x,y)
plt.title('Line chart')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

plt.boxplot(x, vert=False)
plt.title('Box-plot')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()


plt.hist(x, facecolor="Red", alpha=0.5, edgecolor='b')
plt.title('Histogram')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#A3
from matplotlib import pyplot as plt
import numpy as np
subject = ['physics','chemistry','english','maths','biology']
marks =[79, 62, 88, 82, 30]

plt.bar(subject,marks,color='y',width=0.3)
plt.title("BAR CHART")
plt.show()


plt.pie(marks,labels=subject, autopct='%1.1f%%'  )
plt.title("PIE CHART")
plt.show()
```

```python
#A4
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
iris = pd.read_csv("data.csv")
ax=plt.subplots(1,1,figsize=(10,8))
sns.countplot("species",data=iris)
plt.title("Iris Species Count")
plt.show()
```

```python
#A5
import pandas as pd
import matplotlib.pyplot as plt
iris = pd.read_csv("data.csv")
ax=plt.subplots(1,1,figsize=(10,8))
iris['species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.title("Iris Species %")
plt.show()
```

```python
#A6
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
iris = pd.read_csv("data.csv")
plt.hist(iris['species'],facecolor="Grey", alpha=0.5, edgecolor='black')
plt.title("Species")
plt.xlabel("Species type")
plt.ylabel("Characteristic")
plt.show()
```
